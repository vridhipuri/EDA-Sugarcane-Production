```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("/content/drive/MyDrive/List of Countries by Sugarcane Production.csv")
```

```python
df.head(10)
```

```python
df.info()
```

```python
#drop column Unnamed: 0
df.drop("Unnamed: 0",axis=1,inplace=True)
```

```python
df.head()
```

```python
#remove commas and dots from the values
df["Production (Tons)"]=df["Production (Tons)"].str.replace('.','')
df["Production (Tons)"]=df["Production (Tons)"].str.replace(',','')
df["Production per Person (Kg)"]=df["Production per Person (Kg)"].str.replace(',','')
df["Production per Person (Kg)"]=df["Production per Person (Kg)"].str.replace('.','')
df["Acreage (Hectare)"]=df["Acreage (Hectare)"].str.replace('.','')
df["Acreage (Hectare)"]=df["Acreage (Hectare)"].str.replace(',','')
df["Yield (Kg / Hectare)"]=df["Yield (Kg / Hectare)"].str.replace('.','')
df["Yield (Kg / Hectare)"]=df["Yield (Kg / Hectare)"].str.replace(',','')
```

```python
df.head()
```

```python
#convert columns to float
df["Production (Tons)"]=df["Production (Tons)"].astype(float)
df["Production per Person (Kg)"]=df["Production per Person (Kg)"].astype(float)
df["Acreage (Hectare)"]=df["Acreage (Hectare)"].astype(float)
df["Yield (Kg / Hectare)"]=df["Yield (Kg / Hectare)"].astype(float)
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df[df.isnull().any(axis=1)]
```

```python
df.dropna(axis=0, inplace=True)
```

```python
df
```

```python
df.reset_index(drop=True, inplace=True)
```

```python
dfc=df.groupby("Continent").sum()
dfc
```

Which continent produces max sugarcane?

```python
sns.barplot(x="Continent",y="Production (Tons)",data=dfc.sort_values(by="Production (Tons)",ascending=False))
```

How many countries produce sugarcane from each continent?

```python
df["Continent"].value_counts()
```

Which continent has the highest land?

```python
sns.barplot(x="Continent",y="Acreage (Hectare)",data=dfc.sort_values(by="Acreage (Hectare)",ascending=False))
```

*South America has the highest land for the production of sugarcane*

Do number of countries in a Continent effects production of sugarcane?

```python
dfc["no._of_countries"]=df["Continent"].value_counts()
dfc
```

```python
continent_names = dfc.index.to_list()
continent_names
```

```python
sns.lineplot(x="no._of_countries",y="Production (Tons)",data=dfc)
plt.xticks(dfc["no._of_countries"],continent_names,rotation=90)
plt.show()
```
